```python
import pandas as pd
import numpy as np
```

# Load Dataset

```python
df=pd.read_csv(r'/content/spotify-top-50-world.csv')
```

```python
df.shape
```

```python
df.isnull().sum()
```

# Remove Unwanted columns

```python
unwanted_columns = ["temp_column", "duplicate_rank", "random_noise","mostly_null"]
df.drop(columns=unwanted_columns, inplace=True, errors="ignore")
```

```python
df.info()
```

# Handling Missing Values

```python
for col in df.columns:
    if df[col].dtype in ["int64", "float64"]:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df
```

# Remove Duplicate rows

```python
df.drop_duplicates(inplace=True)
```

# Standardize Text Column

```python
text_cols = df.select_dtypes(include="object").columns
for col in text_cols:
    df[col] = df[col].str.strip().str.lower()
```

# Fix Date columns (if any)

```python
for col in df.columns:
    if "date" in col.lower():
        df[col] = pd.to_datetime(df[col], errors="coerce")
```

# Save Cleaned Data

```python
df.to_csv("/content/spotify-top-50-world.csv", index=False)
```
